```python
import pandas as pd
import numpy as np
ds=pd.read_csv('Market_Basket_Optimisation.csv')
ds
```

```python
records =[]
for i in range(0,10):
    records.append([str(ds.values[i,j]) for j in range(2,8)])
```

```python
pip install apyori
```

```python
from apyori import apriori
```

```python
as_rules = apriori(records , min_support = 0.003 ,min_confidence = 0.2 ,min_lift=1.5 , min_length=2)
as_results =list(as_rules)
```
